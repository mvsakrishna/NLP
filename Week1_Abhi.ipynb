# 5. Read bird and cat images, then implement blendImg function (10 points)

````python
# Read and convert images to grayscale
img1 = cv2.imread('/content/drive/MyDrive/Homework1_part2/Homework1_part2/bird.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('/content/drive/MyDrive/Homework1_part2/Homework1_part2/cat.png', cv2.IMREAD_GRAYSCALE)

# Resize the images
img1 = cv2.resize(img1, (480, 320))  # Note: OpenCV uses (width, height)
img2 = cv2.resize(img2, (480, 320))

# Blend the images
alpha = 0.5
beta = 0.5
blended_img = cv2.addWeighted(img1, alpha, img2, beta, 0)

# Display the blended image
plt.imshow(blended_img, cmap='gray')
plt.show()
````

# 7. Read bridge image and add salt & pepper noise (10 points)

````python
# Read the image
img = cv2.imread('/content/drive/MyDrive/Homework1_part2/Homework1_part2/bridge.png', cv2.IMREAD_GRAYSCALE)

# Add salt & pepper noise to the image
noise_img = img.copy()
salt_pepper_prob = 0.02
num_salt = np.ceil(salt_pepper_prob * img.size * 0.5)
x_coords = np.random.randint(0, img.shape[1], int(num_salt))
y_coords = np.random.randint(0, img.shape[0], int(num_salt))
noise_img[y_coords, x_coords] = 255

num_pepper = np.ceil(salt_pepper_prob* img.size * 0.5)
x_coords = np.random.randint(0, img.shape[1], int(num_pepper))
y_coords = np.random.randint(0, img.shape[0], int(num_pepper))
noise_img[y_coords, x_coords] = 0

# Display the noisy image
plt.imshow(noise_img, cmap='gray')
plt.show()
````

# 8. Plot Gaussian function (10 points)

````python
# Define the Gaussian function
def gaussian(x, mu, sig):
    return 1 / (np.sqrt(2 * np.pi * sig**2)) * np.exp(- (x - mu)**2 / (2 * sig**2))

# Generate x values
x = np.linspace(-10, 10, 100)

# Generate y values for a Gaussian with mean 0 and standard deviation 1
y = gaussian(x, 0, 1)

# Plot the Gaussian
plt.figure()
plt.plot(x, y)
plt.title('Gaussian Function')
plt.show()
````

# 9. Add random and Gaussian noise (20 points)

````python
# Read the image
img = cv2.imread('/content/drive/MyDrive/Homework1_part2/Homework1_part2/bridge.png', cv2.IMREAD_GRAYSCALE)

# Add random noise to the image
sigma = 0.1
random_noise = np.random.randn(*img.shape) * sigma
random_noise_img = cv2.normalize(img + random_noise, None, 255,0, cv2.NORM_MINMAX, cv2.CV_8UC1)

# Display the random noise image
plt.imshow(random_noise_img, cmap='gray')
plt.title("Random Noise Image")
plt.show()

# Add Gaussian noise to the image
mean = 0
stddev = 0.1
noise = np.random.normal(mean, stddev, img.shape)
gaussian_noise_img = cv2.normalize(img + noise, None, 255,0, cv2.NORM_MINMAX, cv2.CV_8UC1)

# Display the Gaussian noise image
plt.imshow(gaussian_noise_img, cmap='gray')
plt.title("Gaussian Noise Image")
plt.show()
````
